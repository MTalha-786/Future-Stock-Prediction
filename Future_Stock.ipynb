```python
# =========================
# 1. INSTALL & IMPORT LIBRARIES
# =========================

!pip install yfinance --quiet

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
import warnings
warnings.filterwarnings("ignore")
# Style for plots
sns.set_style("darkgrid")

# =========================
# 2. LOAD DATASET
# =========================
stock_symbol = "TSLA"
df = yf.download(stock_symbol, start="2020-01-01", end="2025-01-01")

# =========================
# 3. BASIC DATA INSPECTION (EDA START)
# =========================

print(" First 5 Rows:")
display(df.head())

print("\n Last 5 Rows:")
display(df.tail())

print("\n Dataset Info:")
print(df.info())

print("\n Columns:")
print(df.columns)

print("\n Missing Values:")
print(df.isnull().sum())

print("\n Statistical Summary:")
display(df.describe())

# =========================
# 4. DATA CLEANING
# =========================

# Keep only important columns
df = df[['Open', 'High', 'Low', 'Close', 'Volume']]

# Check again missing values
df = df.dropna()

print("\n After Cleaning Missing Values:")
print(df.isnull().sum())

# =========================
# 5. FEATURE ENGINEERING
# =========================

# Target = Next day's Close price
df['Target'] = df['Close'].shift(-1)

# Remove last row (NaN target)
df = df.dropna()

# =========================
# 6. DATA VISUALIZATION (EDA)
# =========================

# 6.1 Closing Price Trend
plt.figure(figsize=(12,5))
plt.plot(df['Close'], label="Close Price")
plt.title(" Stock Closing Price Trend")
plt.xlabel("Time")
plt.ylabel("Price")
plt.legend()
plt.show()

# 6.2 Volume Trend
plt.figure(figsize=(12,5))
plt.plot(df['Volume'], color='orange', label="Volume")
plt.title(" Trading Volume Trend")
plt.xlabel("Time")
plt.ylabel("Volume")
plt.legend()
plt.show()

# 6.3 Correlation Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title(" Feature Correlation Heatmap")
plt.show()

# =========================
# 7. SPLIT DATA (TRAIN / TEST)
# =========================

X = df[['Open', 'High', 'Low', 'Close', 'Volume']]
y = df['Target']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=False
)

# =========================
# 8. MODEL TRAINING
# =========================

model = LinearRegression()
model.fit(X_train, y_train)

# =========================
# 9. PREDICTIONS
# =========================

y_pred = model.predict(X_test)

# =========================
# 10. MODEL EVALUATION
# =========================

mae = mean_absolute_error(y_test, y_pred)
print("\n Mean Absolute Error:", mae)

# =========================
# 11. FINAL VISUALIZATION (ACTUAL VS PREDICTED)
# =========================

plt.figure(figsize=(12,6))
plt.plot(y_test.values, label="Actual Price", linewidth=2)
plt.plot(y_pred, label="Predicted Price", linewidth=2)

plt.title(" Actual vs Predicted Stock Prices")
plt.xlabel("Time")
plt.ylabel("Price")
plt.legend()
plt.show()
```
